#### Some values in the columns "build_year", "heating", "furnishment" and "building" are mixed up due to scraping deficiencies. The place of the data were varying between different ads. The strategy to cope with that is to replace meaningless values with "NaN"

#### Checking the unique values of each categorical column

```python
df_copy['street'].nunique()
```

```python
df_copy['build_year'].unique()
```

#### Replacing misvalues with zero in "build_year" column

```python
mask = df_copy['build_year'].str.isalpha()
df_copy.loc[mask, ['build_year']] = "0"
```

```python
df_copy.head()
```

#### Replacing misvalues in "heating" with "0heating"

```python
df_copy['heating'].unique()
```

```python
df_copy1 = df_copy.copy()
```

```python
mask1 = (
    (df_copy1['heating'] == 'Mūrinis') |
    (df_copy1['heating'] == 'Blokinis') |
    (df_copy1['heating'] == 'Pardavimas') |
    (df_copy1['heating'] == 'Medinis') |
    (df_copy1['heating'] == 'Karkasinis') |
    (df_copy1['heating'] == 'Monolitas')
)
df_copy1.loc[mask1, ['heating']] = "0heating"
df_copy1['heating'] = df_copy1['heating'].fillna("0heating")
```

```python
df_copy1['heating'].unique()
```

#### Replacing misvalues in "furnishment" with "0furnishment"

```python
df_copy1['furnishment'].unique()
```

```python
df_copy1.info()
```

```python
df_copy2 = df_copy1.copy()
```

```python
mask2 = (
    (df_copy2['furnishment'] == 'Centrinis') |
    (df_copy2['furnishment'] == 'Pardavimas') |
    (df_copy2['furnishment'] == 'Kietu kuru') |
    (df_copy2['furnishment'] == 'Dujinis') |
    (df_copy2['furnishment'] == 'Centrinis-kolektorinis') |
    (df_copy2['furnishment'] == 'Mūrinis') |
    (df_copy2['furnishment'] == 'Blokinis') |
    (df_copy2['furnishment'] == 'Elektrinis') |
    (df_copy2['furnishment'] == 'Autonominis') |
    (df_copy2['furnishment'] == 'Monolitas')
)
df_copy2.loc[mask2, ['furnishment']] = "0furnishment"
df_copy2['furnishment'] = df_copy2['furnishment'].fillna("0furnishment")
```

```python
df_copy2.head()
```

```python
df_copy2['furnishment'].unique()
```

#### Replacing misvalues in "building" with "0building"

```python
df_copy2['building'].unique()
```

```python
df_copy2.info()
```

```python
df_copy3 = df_copy2.copy()
```

```python
mask3 = (df_copy3['building'] == 'Pardavimas') 
df_copy3.loc[mask3, ['building']] = "0building"
df_copy3['building'] = df_copy3['building'].fillna("0building")
```

#### Flat price Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
df1 = pd.read_csv('data.csv')
```

```python
features = [
        'city_numeric',
        'street_numeric',
        'area',
        'number_of_rooms',         
        'floors',                  
        'floor',                   
        'price_per_square_m',                  
        'build_year',             
        '0heating',                 
        'Aeroterminis',            
        'Autonominis',              
        'Centrinis',                
        'Centrinis-kolektorinis',   
        'Dujinis',                 
        'Elektrinis',               
        'Kietu kuru',              
        'Kitas',
        '0furnishment',                                    
        'Dalinė apdaila',                   
        'Neįrengtas / Statomas',             
        'Pilnai įrengtas',         
        'Reikalingas remontas',    
        'Suremontuotas',           
        'Tvarkingas',             
        '0building',
        'Blokinis',           
        'Karkasinis',              
        'Medinis',                 
        'Monolitas',               
        'Mūrinis'                  
]

label = ['total_price']
```

```python
x = df1[features].values
y = df1[label]
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
print(type(x_train))
print(type(x_test))
print(type(y_train))
print(type(y_test))
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print(y_test[0])
print(y_pred[0])
```

```python
type(y_test)
```

```python
print("MAE", mean_absolute_error(y_test, y_pred))
print("MSE", mean_squared_error(y_test, y_pred))
```

```python
explained_variance_score(y_test, y_pred)
```
